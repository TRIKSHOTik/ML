```python
from sklearn.preprocessing import LabelEncoder, TargetEncoder
from sklearn.feature_extraction import FeatureHasher

print("ДО:")
print(df['brand'].value_counts())
print(f"Всего брендов: {df['brand'].nunique()}")
df_encoded = pd.get_dummies(df, columns=['brand'], prefix='brand')
print("\nПОСЛЕ:")
brand_cols = [col for col in df_encoded.columns if col.startswith('brand_')]
print(brand_cols[:10])
```

```python
median_price = df['price'].median()
df['target_high_price'] = (df['price'] > median_price).astype(int)

target_enc = TargetEncoder(target_type='binary')
df['brand_target_encoded'] = target_enc.fit_transform(
    df['brand'].fillna('Unknown').to_frame(),
    df['target_high_price']
)

result = df[['brand', 'brand_target_encoded']].drop_duplicates().sort_values('brand_target_encoded', ascending=False)
print(result.head(10))
```

```python
descriptions = df['description'].fillna('').head(1000).tolist()
feature_dicts = [{'desc': d} for d in descriptions]
hasher = FeatureHasher(n_features=10, input_type='dict')
hashed_features = hasher.transform(feature_dicts).toarray()
hashed_df = pd.DataFrame(hashed_features, columns=[f'desc_hash_{i}' for i in range(10)])
print("Хэшированные признаки:")
print(hashed_df.head())
```

```python
df['final_price'] = df['list_price'].fillna(df['price'])
df['discount_amount'] = df['list_price'] - df['price']
df['discount_amount'] = df['discount_amount'].clip(lower=0)
df['price_per_star'] = df['price'] / df['stars'].clip(lower=0.1)
print(df[['price', 'list_price', 'final_price', 'discount_amount', 'price_per_star']].head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import plotly.graph_objects as go
from plotly.subplots import make_subplots

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

df_clean = df.dropna(subset=['price', 'stars', 'brand']).copy()
df_clean['discount_pct'] = pd.to_numeric(df_clean['discount_pct'], errors='coerce')
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(df_clean['price'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title('Распределение цен (гистограмма)', fontsize=14)
axes[0].set_xlabel('Цена ($)', fontsize=12)
axes[0].set_ylabel('Частота', fontsize=12)
axes[0].axvline(df_clean['price'].mean(), color='red', linestyle='--', label=f'Среднее: ${df_clean["price"].mean():.0f}')
axes[0].axvline(df_clean['price'].median(), color='green', linestyle='--', label=f'Медиана: ${df_clean["price"].median():.0f}')
axes[0].legend()

sns.kdeplot(data=df_clean, x='price', fill=True, color='purple', alpha=0.5, ax=axes[1])
axes[1].set_title('Плотность распределения цен (KDE)', fontsize=14)
axes[1].set_xlabel('Цена ($)', fontsize=12)
axes[1].set_ylabel('Плотность', fontsize=12)
axes[1].axvline(df_clean['price'].mean(), color='red', linestyle='--', label=f'Среднее')
axes[1].axvline(df_clean['price'].median(), color='green', linestyle='--', label=f'Медиана')
axes[1].legend()

plt.tight_layout()
plt.show()

print("\n📊 Анализ распределения цен:")
print(f"Средняя цена: ${df_clean['price'].mean():.2f}")
print(f"Медианная цена: ${df_clean['price'].median():.2f}")
print(f"Стандартное отклонение: ${df_clean['price'].std():.2f}")
print(f"Минимальная цена: ${df_clean['price'].min():.2f}")
print(f"Максимальная цена: ${df_clean['price'].max():.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].scatter(df_clean['stars'], df_clean['price'],
                alpha=0.5, c='blue', edgecolors='black', linewidth=0.5)
axes[0].set_title('Зависимость цены от рейтинга', fontsize=14)
axes[0].set_xlabel('Рейтинг (звезды)', fontsize=12)
axes[0].set_ylabel('Цена ($)', fontsize=12)
axes[0].grid(True, alpha=0.3)

mean_price = df_clean['price'].mean()
axes[0].axhline(y=mean_price, color='red', linestyle='--',
                label=f'Средняя цена: ${mean_price:.0f}')
axes[0].legend()

scatter = axes[1].scatter(df_clean['stars'], df_clean['price'],
                          c=pd.factorize(df_clean['brand'])[0],
                          cmap='viridis', alpha=0.6, edgecolors='black', linewidth=0.5)
axes[1].set_title('Цветовая дифференциация по брендам', fontsize=14)
axes[1].set_xlabel('Рейтинг (звезды)', fontsize=12)
axes[1].set_ylabel('Цена ($)', fontsize=12)
plt.colorbar(scatter, ax=axes[1], label='Бренд (код)')
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=mean_price, color='red', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

fig = px.scatter(df_clean, x='stars', y='price', color='brand',
                 hover_data=['title', 'discount_pct'],
                 title='Интерактивный scatter plot: Цена vs Рейтинг',
                 labels={'stars': 'Рейтинг', 'price': 'Цена', 'brand': 'Бренд'},
                 trendline='ols')
fig.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].boxplot(df_clean['price'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='lightblue'),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'),
                flierprops=dict(marker='o', markerfacecolor='red', markersize=8, alpha=0.5))
axes[0].set_title('Box plot цен (matplotlib)', fontsize=14)
axes[0].set_ylabel('Цена ($)', fontsize=12)
axes[0].set_xticklabels(['Цена'])
axes[0].grid(True, alpha=0.3)

sns.boxplot(data=df_clean, x='brand', y='price', ax=axes[1])
axes[1].set_title('Распределение цен по брендам (seaborn)', fontsize=14)
axes[1].set_xlabel('Бренд', fontsize=12)
axes[1].set_ylabel('Цена ($)', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

sns.violinplot(data=df_clean, x='brand', y='price', ax=axes[2])
axes[2].set_title('Violin plot - распределение цен', fontsize=14)
axes[2].set_xlabel('Бренд', fontsize=12)
axes[2].set_ylabel('Цена ($)', fontsize=12)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

fig = px.box(df_clean, x='brand', y='price', color='brand',
             title='Интерактивный box plot: Распределение цен по брендам',
             points='all',
             hover_data=['title', 'stars'])
fig.show()

print("\n🔍 Анализ выбросов:")
Q1 = df_clean['price'].quantile(0.25)
Q3 = df_clean['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[(df_clean['price'] < lower_bound) | (df_clean['price'] > upper_bound)]
print(f"Количество выбросов: {len(outliers)}")
print(f"Процент выбросов: {len(outliers)/len(df_clean)*100:.1f}%")
print(f"Границы: [${lower_bound:.0f}, ${upper_bound:.0f}]")
if len(outliers) > 0:
    print("\nПримеры выбросов (самые дорогие):")
    print(outliers.nlargest(5, 'price')[['title', 'brand', 'price', 'stars']])
```

```python
brand_counts = df_clean['brand'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].bar(brand_counts.index, brand_counts.values, color='skyblue', edgecolor='black')
axes[0].set_title('Топ-10 брендов по количеству моделей (matplotlib)', fontsize=14)
axes[0].set_xlabel('Бренд', fontsize=12)
axes[0].set_ylabel('Количество моделей', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
for i, v in enumerate(brand_counts.values):
    axes[0].text(i, v + 0.5, str(v), ha='center', fontsize=10)

sns.countplot(data=df_clean, y='brand', order=df_clean['brand'].value_counts().index[:10], ax=axes[1])
axes[1].set_title('Распределение моделей по брендам (seaborn)', fontsize=14)
axes[1].set_xlabel('Количество', fontsize=12)
axes[1].set_ylabel('Бренд', fontsize=12)

plt.tight_layout()
plt.show()

fig = px.bar(x=brand_counts.index, y=brand_counts.values,
             title='Интерактивный bar chart: Количество моделей по брендам',
             labels={'x': 'Бренд', 'y': 'Количество моделей'},
             color=brand_counts.values,
             color_continuous_scale='Viridis')
fig.show()

fig = px.pie(values=brand_counts.values, names=brand_counts.index,
             title='Доля рынка по брендам (топ-10)',
             hole=0.3)
fig.show()
```

```python
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df_clean[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Корреляционная матрица числовых признаков', fontsize=16)
plt.tight_layout()
plt.show()

fig = px.imshow(correlation_matrix,
                text_auto='.2f',
                aspect="auto",
                color_continuous_scale='RdBu_r',
                title='Интерактивная корреляционная матрица')
fig.update_layout(width=800, height=700)
fig.show()

print("\n📈 АНАЛИЗ КОРРЕЛЯЦИЙ:")
print("-" * 40)

strong_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr = correlation_matrix.iloc[i, j]
        if abs(corr) > 0.5:
            strong_correlations.append((correlation_matrix.columns[i],
                                       correlation_matrix.columns[j],
                                       corr))

if strong_correlations:
    print("Сильные корреляции (> 0.5 или < -0.5):")
    for col1, col2, corr in sorted(strong_correlations, key=lambda x: abs(x[2]), reverse=True):
        print(f"  • {col1} и {col2}: {corr:.3f}")
else:
    print("Сильных корреляций не обнаружено")
```
